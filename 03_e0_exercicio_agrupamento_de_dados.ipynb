## Solução


### Carga do conjunto de dados

```python
import pandas as pd
from sklearn.datasets import load_iris

iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['target'] = iris.target
X = df.drop('target', axis=1)

#y = df['target']

df.head()
```

### K-means

```python
for i in range(1,5):
  for j in range(2,5):
    if j> i:
      print(i,j)
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

number_clusters = [2,3,6]

for k in number_clusters:
  kmeans = KMeans(n_clusters=k)
  kmeans.fit(X)

  for i in range(1,5):
    for j in range(2,5):
      if j> i:
        plt.scatter(X.iloc[:, i], X.iloc[:, j], c=kmeans.labels_, cmap='viridis', label='Clusters')
        plt.scatter(kmeans.cluster_centers_[:, i], kmeans.cluster_centers_[:, j], s=200, c='red', label='Centroids', marker='X')
        plt.legend()
        plt.show()

```
